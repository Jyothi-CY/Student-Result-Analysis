```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/dataset_FDS/Student_score_analysis.csv")
```

```python
# Checking for null values
df.isnull().sum()
```

```python
#From the data set to analyse we need main columns like math score, reading score and writing score
# Here there are no null values in these columns so, no need to fill the values here.

# So, we can delete unnamed columns here.
```

```python
#Drop unnamed column

df= df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
# Changing weekly study column

df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
# Count the values

#Gender Distribution
plt.figure(figsize=(4,4))
ax = sns.countplot(data= df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore": "mean","WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

```python
# Darker shades - Lower marks
# Lighter shades - higher marks
```

```python
# Consider now you don't have parent education here, but you want to see the parent's marital status, that if they are married, if their parents are
# married, if they are single, if they are divorced, if you want to see if it has any impact on the students' scores, then again 
# this time I am going to do group by again, but this time I will do group by on the basis of the parent's marital status.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore": "mean","WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's score")
plt.show()
```

```python
# We can conclude that from above chart, that there is no/negligible impact on the students score due to their parent's marital status.
```
